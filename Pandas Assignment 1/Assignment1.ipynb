## 1. Read only time, geo and value columns <br><br> 2. Consider “:” as missing

```python
data1  = pd.read_csv("data.csv",usecols = ['TIME','GEO','Value'],na_values=':')
data1.shape
```

## 3. Fill the missing data appropriately

```python
data1.isnull().sum()
```

```python
data1.count()
```

```python
data1.plot()
```

```python
data1.fillna(data1['Value'].mean()).plot()
```

```python
data1.fillna(data1['Value'].mean(),inplace= True)
data1.isnull().count()
```

## 5. Select all rows where Value is greater than 5 and year is 2011

```python
data1[(data1.Value > 5) & (data1.TIME == 2011)]
```

## 6. Display max value of all columns

```python
data1.max()
```

## 7. Create a new column val_sq  with square of the value column in it

```python
data1['val_sq'] = data1.Value ** 2
data1.sample(20)
```

## 8. Delete newly added column

```python
data1.drop('val_sq',axis=1,inplace=True)
```

```python
data1.head()
```

## 9.Add a new value 2000,  5.00, 'a'  to the end of Dataframe

```python
new = pd.Series({'TIME':2000,
                    'GEO':'a',
                    'Value':5.00})
new
```

```python
data1 = data1.append(new,ignore_index=True)
data1.tail()
```

## 10.Sort the dataframe in the descending order of values

```python
data1.sort_values('Value',ascending = False)
```

## 11. Show the mean of the values for each country over all the years

```python
data1.groupby(['GEO']).Value.mean()
```
